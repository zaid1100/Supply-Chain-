# Sum of total Sales by Customer Segment

```python
total_sales = df.groupby('Customer Segment')['Sales'].sum().reset_index()

# convert sales in millions
total_sales['Sales'] = total_sales['Sales'] / 1_000_000

# create a barplot
ax=sns.barplot(data=total_sales,x='Customer Segment',y='Sales')
for bars in ax.containers:
    ax.bar_label(bars,fmt='%.2fM')
plt.ylabel('Sales In Millions')
plt.xlabel('Names Of The Customer Segment')
plt.title('Total Sales By Customer Segment Wise')
```

# Top 10 Category With Maximum Sales

```python
plt.figure(figsize=(12,5))
# Group by Category Name and sum Sales
top_sales = df.groupby('Category Name')['Sales'].sum().reset_index()

# Sort by Sales in descending order and get the top 10 categories
top_sales_sorted = top_sales.sort_values(by='Sales', ascending=False).head(10)
# Convert Sales values to millions
top_sales_sorted['Sales'] = top_sales_sorted['Sales'] / 1_000_000

# Plot
ax = sns.barplot(data=top_sales_sorted, x='Category Name', y='Sales')

# Label the bars with values in millions
for bars in ax.containers:
    ax.bar_label(bars, fmt='%.1fM')

plt.xticks(rotation=45)  
plt.xlabel('Category Name')  
plt.ylabel('Sales (Millions)')  
plt.title('Top 10 Categories by Sales')
```

# Count Top 10 Cities Maximum Customers 

```python
plt.figure(figsize=(12,4))
# count of maximum customer used group by on customer city 
customer_df = df.groupby('Customer City')['Customer Id'].count().reset_index()

# sort values in ascending orders by customer id
sort_values=customer_df.sort_values(by='Customer Id',ascending=False).head(10)

# created a barplot
ax=sns.barplot(data=sort_values,x='Customer City',y='Customer Id')
for bars in ax.containers:
    ax.bar_label(bars)
plt.xlabel('Names Of The Cities')
plt.ylabel('Count Of Customer Id')
plt.title('Top Cities Count Maximum Customer In Which City')
```

# Which products generate the highest and lowest sales, and how do they perform in terms of profitability?

```python
# Top 10 products by sales
top_products_by_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
# Top 10 products by profit
top_products_by_profit = df.groupby('Product Name')['Order Profit Per Order'].sum().sort_values(ascending=False).head(10)

# Plotting side by side bar plots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar plot for top 10 products by sales
sns.barplot(x=top_products_by_sales.values, y=top_products_by_sales.index, ax=axes[0], palette='Blues_d')
axes[0].set_title('Top 10 Products by Sales')
axes[0].set_xlabel('Total Sales')

# Bar plot for top 10 products by profit
sns.barplot(x=top_products_by_profit.values, y=top_products_by_profit.index, ax=axes[1], palette='Greens_d')
axes[1].set_title('Top 10 Products by Profit')
axes[1].set_xlabel('Total Profit')

plt.tight_layout()
plt.show()
```

# Which states/cities have the highest sales growth over time?

```python
# Group sales by state and order date to calculate sales over time
df['Order YearMonth'] = df['order date'].dt.to_period('M')

sales_trend_by_state = df.groupby(['Order State', 'Order YearMonth'])['Sales'].sum().unstack().fillna(0)
print(sales_trend_by_state)

# Visualization for a particular state (e.g., 'California')
sales_trend_by_state.loc['California'].plot(kind='line', title='Sales Growth in California Over Time')
plt.ylabel('Sales')
plt.xlabel('Month')
plt.show()
```

# What is the overall impact of delayed orders on profitability?
